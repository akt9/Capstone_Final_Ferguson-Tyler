### step 1: finding the missing exectives within the company records 

we want to start by finding the names from article 460 in the employee database to find out who they are and who they were interacting with

names to filter the dataframe on were taken from article 460 naming the missing executives, the first list was a bit exacting so we tried out some last only combinations to find the persons in question

```python
missing_execs = ['Sten Sanjorge', 'Ingrid Precipice', 'Field-Corrente' , 'Rasgueo of Orhan', 'Willem Basoue-Country']
```

```python
missing_execs_lastOnly = ['Sanjorge', 'Ingrid', 'Precipice', 'Corrente' , 'Rasgueo', 'Orhan', 'Willem', 'Basoue-Country']
```

finally, pulling in the employee records

```python
try:
    top_level = pd.read_excel('C:/Users/Andy/PycharmProjects/pythonProject/The_Last_Stand/TP-1_Kronos/EmployeeRecords.xlsx')
except:
    top_level = pd.read_excel('C:/Users/andyt/PycharmProjects/classworker_General/The_Last_Stand/TP-1_Kronos/EmployeeRecords.xlsx')

display(top_level.head())
```

so using the list from we recalculated above, we find that the misisng persons all seem to be high level executives at GasTech

```python
mask = top_level['LastName'].str.contains(missing_execs_lastOnly[0])
for val in missing_execs_lastOnly[1:]:
    mask |= top_level['LastName'].str.contains(val)
    mask |= top_level['FirstName'].str.contains(val)
filtered_top_level = top_level[mask]
display(filtered_top_level)
```

names and employment titles are consistent with the names in the article. although the data doesn't match 1:1, job titles, number of returns, and no missing or plural entries would indicate that these are our missing persons, and they happen to be the executives. this lines up with a PoK plot.

### step 2: looking at the email traffic of the entity emails

so now we need to start identifying the other nodes, by creating a small frame with the executives only, we can cross reference the email addresses from the employee records frame with the email headers and find all traffic involving the executives

next we pull in the email traffic database and filter it using an inner join of the emails in the victims frame

```python
try:
    headers = pd.read_csv('C:/Users/Andy/PycharmProjects/pythonProject/The_Last_Stand/TP-1_Kronos/email_headers.csv', encoding='cp1252')
except:
    headers = pd.read_csv('C:/Users/andyt/PycharmProjects/classworker_General/The_Last_Stand/TP-1_Kronos/email_headers.csv', encoding='cp1252')


display(headers.head())
```

```python
mask = headers['From'].isin(filtered_top_level['EmailAddress'])
outgoing_df = headers[mask]

outgoing_df['To'] = outgoing_df['To'].str.split(',')

display(outgoing_df)
```

having done this, we now need to identify all persons that received an email from one of our executives, we suspect that some of these persons may be suspects, or have one hop connections to the executives as they may have been on the receiving end of the information from someone in contact with the victims

```python
receivers = outgoing_df.explode('To')
receivers = receivers['To'].drop_duplicates().to_list()
receivers = list(set(receivers)-set(filtered_top_level['EmailAddress'].to_list()))
print(receivers)
print(len(receivers))
```

```python
receivers_df = top_level[top_level['EmailAddress'].isin(receivers)]
display(receivers_df)
print(receivers_df.shape)
```

not everyone who was sent an email by one of our original 5 is in the employee records, but this is a start

### step 3: people who sent emails to the original 5

now we perform the same operation using people who may have sent an email to the victims, maybe they told them to meet somewhere? it also just helps flesh out the network.


```python
incoming_df = headers.copy()
incoming_df['To'] = incoming_df['To'].str.split(',')

mask = incoming_df['To'].apply(lambda x: any(item in x for item in filtered_top_level.EmailAddress.to_list()))
filtered_senders = incoming_df[mask]

display(filtered_senders)
```

```python
senders_df = top_level[top_level['EmailAddress'].isin(filtered_senders['From'])]
display(senders_df)
print(senders_df.shape)
```

people who sent an email to one of the five victims


### step 3: combining both lists to generate a list of suspects

finally, we will combine these two dataframes to form what is ostensibly our list of nodes

```python
suspects_df = pd.concat([receivers_df, senders_df]).drop_duplicates()

suspects_df = suspects_df[~suspects_df['EmailAddress'].isin(filtered_top_level.EmailAddress.to_list())]

display(suspects_df)
```

off the bat, these two look suspicious, both kronosian, both about the same age, both with armed services histories and left on the same day, we're going to pay special attention to email traffic and articles that mention these too
