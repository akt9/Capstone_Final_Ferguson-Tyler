```python
# Importing the employees file 
employee = pd.read_csv(current_directory+'\\TP-2_Kronos\\EmployeeRecords.csv', encoding='cp1252')
employee['FullName'] =  employee['FirstName'] + ' ' + employee['LastName']
employee.head()
```

```python
# Combining the employees file with the cc and gps and loyalty data
df = pd.merge(cc_gps_lty, employee, how='left', on='FullName').copy()

df['CurrentEmploymentType'] = df['CurrentEmploymentType_x']
df.head()
```

```python
df.columns
```

```python
# Merging together the current employment type
def combine_employment_types(row):
    if row['CurrentEmploymentType_x'] == row['CurrentEmploymentType_y']:
        return row['CurrentEmploymentType_x']
    elif row['CurrentEmploymentType_x'] != 'Unknown':
        return row['CurrentEmploymentType_x']
    else:
        return row['CurrentEmploymentType_y']

def combine_employment_title(row):
    if row['CurrentEmploymentTitle_x'] == row['CurrentEmploymentTitle_y']:
        return row['CurrentEmploymentTitle_x']
    elif row['CurrentEmploymentTitle_x'] != 'Unknown':
        return row['CurrentEmploymentTitle_x']
    else:
        return row['CurrentEmploymentTitle_y']


# Apply the function to create the new column
df['CurrentEmploymentType'] = df.apply(combine_employment_types, axis=1)
df['CurrentEmploymentTitle'] = df.apply(combine_employment_title, axis=1)
```

```python
df = df.drop(columns=['CurrentEmploymentType_x', 'CurrentEmploymentTitle_x', 'CurrentEmploymentType_y','CurrentEmploymentTitle_y'])
```

```python
df['lat'] =  df['lat'].astype('float64')
df['long'] = df['long'].astype('float64')
df['day'] =  df['day'].astype('Int32')
df['hour'] =  df['hour'].astype('Int32')
df['minute'] =  df['minute'].astype('Int32')
```

```python
df['float_lat'] = [float(i) for i in df['lat']]
df['float_long'] = [float(i) for i in df['long']]
df = df.drop(columns=['lat','long'])

# Exporting out the df full sheet to be used with tableau
df.to_csv('df_full.csv')
```
